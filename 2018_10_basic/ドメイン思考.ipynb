### ドメインモデルとは

ドメインモデルは業務の関心事を表現するドメインオブジェクトを集めて体系的に整理したものです。ドメインオブジェクトは業務の関心ごとに直接的に対応します





クラス名やメソッド名は業務の用語そのものです。
それぞれのドメインオブジェクトは業務データとそれを使った判断加工計算のロジックを一体にしたものです。データとロジックを一つのクラスにまとめることで業務ルールに基づく判断/加工/計算のコードの重複を防ぎます

### ドメインモデルを 設計する狙いは次の三つ。
- 業務的な判断化工計算のロジックを重複なく一元的に記述する
- 業務の関心事と行動を直接対応させどこに何が書いてあるかわかりやすく整理する
- 業務ルールの変更や追加の時に変更の影響を狭い範囲に閉じ込める

<b>ドメインモデルは プログラム言語で書いた業務の用語集であり業務の説明書です</b>  
それぞれの用語がどのように関連しどのように総合に作用する顔パッケージ構成やクラスの参照関係で立体的に表現する手段です。  
業務の関心ごとの単位とプログラムの単位が一致していると業務ルールや修正や追加が楽で安全になりますどこに何が書いてわかりやすく修正の対象が一箇所にまとまります。  
プログラムを変更した時に業務的に関係のない箇所に思わぬ副作用が起きなくなります。

### 業務に使っている用語をクラス名にする

ドメインモデルの設計は業務で使われる具体的な用語かっこ概念を手がかりに進めますそしてその用語がデータアートロジックを一塊としたプログラム単位として使えそうなことを検証します そうやって業務の関心ごと教務で使われている用語を理解しながらプログラムの構造を考えていくのがオブジェクト指向の分析設計のやり方です

<b>ドメインオブジェクトの設計の基本は現実の業務の中で使われている具体的な言葉の単位で業務ロジックを整理することです</b>

クラス名やメソッド名を業務で使う用語の単位と一致させる設定をまず行いそれだけではうまく整理できない部分が明らかになったらそこに集中してもっとうまく整理できる言葉
<hr>

### 設計時のありがちな失敗は
ドメインモデルの設計のありがちな失敗に業務では実際には使っていない抽象的な言葉をクラス名にして使ってしまうことがあります。  
クラス名を抽象的にすればするほどその名前は広い範囲の対象を包囲して説明できます。  
抽象的で意味の広い名前をクラス名やパッケージ名にした方がさまざまな要素をシンプルにすっきりに整理できるような錯覚しがちですが,そういう意味の広い中周期的な名前を使ったクラスは<b>具体的に何も説明していません。</b>

業務の現実の詳細を的確に捉えていないのです。 

#### 例えば  
業務の様々な活動をどれも取引として説明することはできます販売も仕入れも取引の一種です。 
だからといって取引クラスに販売の業務ルールも、仕入れの業務ルールもどちらも記述するのはプログラムを複雑にするだけです。
やめておきましょう。
  
なんだけど・・・・・・
業務の用語として使われていない抽象的な概念をの導入が分析家設計にブレイクスルーをもたらすことがあります。ごちゃごちゃしてわかりにくかったクラスのもつれた関係が抽象的な概念を表現するクラスの発見によりすっきりと整理することもあります。しかし、業務の現場で使っていない抽象概念をクラス名に採用してしまう でうまくいくのはドメインモデルの中ではごく一部です

### ドメインモデルをどう作るか

#### ①パターン　手続き型アプローチ
全体を俯瞰し定義するところからスタートします。  
段階的に分割しながらより詳細に定義を進めます。   
その時に分割を同じレベルで進めることを重視します初期の段階では全ての粒度が大きく中間地点では全ての要素が10ぐらいの粒度になります。 
最終的に全てをプログラミング可能な小さな粒度で定義します。
いわゆる<b>トップダウンのアプローチ</b>です分割のそれぞれの段階で全体的に網羅性や整合性を確認しながら詳細化をしていくやり方です

#### ②パターン　オブジェクト指向アプローチ
部分に注目します。個々の部品を作り始めそれを組み合わせながら段階的に全体を作っていきます。ボトムアップのアプローチです全体を定義し網羅性や整合性を重視しながら部分を設計していくとトップダウンのアプローチは データとロジックの一緒に考えるオブジェクト指向のやり方ではうまくいきません。
全体を俯瞰し段階的に分解していくのはそれだけでも大変な作業です。  
その上データとロジックの詳細まで一緒に考えるのは負担が大きく非効率です。どうしても全体から部分を分割式たいのであればデータはデータ機能は昨日に分けて段階的に分割している方がまだやるやりやすいでしょう 。

一方部分から全体を組み立てよう組み立てていくボトムアップのアプローチであれば、データとロジックを一緒に考えることはむしろ自然ですある部分に注目してその部分に必要なデータとそのデータを使った判断学校設計のロジックだけを範囲を限定すれば、データアートロジックを一緒に考えそのままクラスとして設定することは難しいことではありませんそうやって部分に注目してデータの業務ロジックを一体にして小さな部分ドメインオブジェクトを設定しながら全体を組み合わせていくのがオブジェクト指向のやり方です

### 全体と部分を行ったり来たりしながら作っていく

部分に注目するボトムアップ型のオブジェクト指向のアプローチでもちろんー全体を意識して開発します全体像を無視してどこでもいいからとにかく細部から手をつけるわけでありません。部分だけに目を向けていると全体として間違った方向に進む危険性がありますそうならないために全体的に俯瞰していつも意識します


### 全体を俯瞰する便利な図  
- パッケージ図
- 業務フロー図

### パッケージ図
個々のクラスを隠蔽しパッケージ単体で全体の構造を俯瞰する手段です 業務の理解が深まり実装されたクラスが増えるに従いパッケージの構成や名前を改善していきます。  
分析の初期の段階はクラス単位で考えるよりも業務ロジックでおおよその置き場所をパッケージとして割り振ってみます。  
パッケージ図はパッケージ間の依存関係を表現できます業務アプリケーションの場合パッケージの依存関係は業務の前後関係と一致します。  
パッケージからの矢印の方向はオブジェクトを参照する方向です。  
受注オブジェクトは顧客オブジェクトと商品オブジェクトと在庫オブジェクトを参照します。  
これは業務の流れの時間軸を逆に辿った関係です。  
業務フローの最初に登場するオブジェクトは後から登場するオブジェクトを知ることがありません。  
業務の進行に伴って後から登場するオブジェクトが先行するオブジェクトを参照しますこのように業務の流れに沿った時間軸がオブジェクト間にいい関係になります。  
パッケージ間の参照関係は次の説明する業務フローを描きながら業務の流れの中の関係として説明します。

### ドメインオブジェクトの見つけ方

業務に必要なすべてのドメインオブジェクトを最初から全て見つけることはできません。
ドメインモデルに本来あるべき部品が足りないことは機能を組み立てていく過程で明らかになり、そのように不足部分を見つけながら、ドメインモデルに追加していくのが業務アプリケーションを設計していくやり方です。 

業務アプリケーションは 対象とする業務データの種類が多くのデータを使った様々な処理（判断・加工・計算）の業務ロジックを実現するには多くの業務知識が必要です。  
アプリケーション全体を設計するために大量の業務知識を相手にして重要な点とそうでない点を判別し表面的には理解し難い背後にある重要な関係性や構造を発見して行くにはどうすれば良いでしょうか

### 業務の関心事をヒト・モノ・コトの三つに分類する

#### ヒト
業務活動の当事者で、意思があり、判断をし、行動する主体です。   
ヒトという関心ごとに対応するドメインオブジェクトは<b>ヒトの意思・判断・行動についてのデータ</b>を持ちそのデータを使った<b>判断・加工・計算のロジック</b>を持ちます

#### モノ

ヒトが業務を遂行するときの関心の対象です。物理的なものだけではなく権利や義務のように概念的に認識している対象もです。  
- 数量、金額、率
- 説明、注釈
- 状態
- 日月や期間
- 位置

これらの属性を表現するデータとそのデータに対して業務的にはどういう<b>判断・加工・計算をしたかをロジック</b>として持ちます

#### コト
業務活動では様々なことが起こります。業務アプリケーションの基本的な関心事はコトのを記憶しコトの発生を通知することです。  
ことは基本的にヒトの意思決定や行動の結果です。  

ことを表現するオブジェクトが持つ一般的な属性は次の通りです。  
- 対象（何について発生した事象か）
- 種別（どう言う種類の事象か）
- 時点（いつ起きた事象か）

このような業務的に関心のある事象を適切に記録し管理することが業務アプリケーションの基本です。この事象が業務的に起きて良いことなのかの判断や起きた場所に何をすべきかの判断を重要な業務の約束事です

#### コトに注目する。
<b>-コトはヒトとモノとの関係として出現する</b>  
<b>-コトは時間軸に沿って明確な前後関係をもつ</b>
