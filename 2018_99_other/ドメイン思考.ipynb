### ドメインモデルとは

ドメインモデルは業務の関心事を表現するドメインオブジェクトを集めて体系的に整理したものです。ドメインオブジェクトは業務の関心ごとに直接的に対応します





クラス名やメソッド名は業務の用語そのものです。
それぞれのドメインオブジェクトは業務データとそれを使った判断加工計算のロジックを一体にしたものです。データとロジックを一つのクラスにまとめることで業務ルールに基づく判断/加工/計算のコードの重複を防ぎます

### ドメインモデルを 設計する狙いは次の三つ。
- 業務的な判断化工計算のロジックを重複なく一元的に記述する
- 業務の関心事と行動を直接対応させどこに何が書いてあるかわかりやすく整理する
- 業務ルールの変更や追加の時に変更の影響を狭い範囲に閉じ込める

<b>ドメインモデルは プログラム言語で書いた業務の用語集であり業務の説明書です</b>  
それぞれの用語がどのように関連しどのように総合に作用する顔パッケージ構成やクラスの参照関係で立体的に表現する手段です。  
業務の関心ごとの単位とプログラムの単位が一致していると業務ルールや修正や追加が楽で安全になりますどこに何が書いてわかりやすく修正の対象が一箇所にまとまります。  
プログラムを変更した時に業務的に関係のない箇所に思わぬ副作用が起きなくなります。

### 業務に使っている用語をクラス名にする

ドメインモデルの設計は業務で使われる具体的な用語かっこ概念を手がかりに進めますそしてその用語がデータアートロジックを一塊としたプログラム単位として使えそうなことを検証します そうやって業務の関心ごと教務で使われている用語を理解しながらプログラムの構造を考えていくのがオブジェクト指向の分析設計のやり方です

<b>ドメインオブジェクトの設計の基本は現実の業務の中で使われている具体的な言葉の単位で業務ロジックを整理することです</b>

クラス名やメソッド名を業務で使う用語の単位と一致させる設定をまず行いそれだけではうまく整理できない部分が明らかになったらそこに集中してもっとうまく整理できる言葉

### 設計時のありがちな失敗は
ドメインモデルの設計のありがちな失敗に業務では実際には使っていない抽象的な言葉をクラス名にして使ってしまうことがあります。  
クラス名を抽象的にすればするほどその名前は広い範囲の対象を包囲して説明できます。  
抽象的で意味の広い名前をクラス名やパッケージ名にした方がさまざまな要素をシンプルにすっきりに整理できるような錯覚しがちですが,そういう意味の広い中周期的な名前を使ったクラスは<b>具体的に何も説明していません。</b>

業務の現実の詳細を的確に捉えていないのです。 

#### 例えば  
業務の様々な活動をどれも取引として説明することはできます販売も仕入れも取引の一種です。 
だからといって取引クラスに販売の業務ルールも、仕入れの業務ルールもどちらも記述するのはプログラムを複雑にするだけです。
やめておきましょう。
  
なんだけど・・・・・・
業務の用語として使われていない抽象的な概念をの導入が分析家設計にブレイクスルーをもたらすことがあります。ごちゃごちゃしてわかりにくかったクラスのもつれた関係が抽象的な概念を表現するクラスの発見によりすっきりと整理することもあります。しかし、業務の現場で使っていない抽象概念をクラス名に採用してしまう でうまくいくのはドメインモデルの中ではごく一部です

#### ドメインモデルは 業務ロジックの整理の手法です。

業務データを判断加工計算するための業務ロジックを データとひとまとまりにしてクラスという単位で整理するのがオブジェクト指向の考え方で関心の中心は業務ロジックでありデータではありません。

#### データモデルは文字通りデータが主役です。

業務で発生する様々なデータを整理してどうテーブルに記録するかを考えます。


### 生年月日と年齢を例にしてドメインモデルとデータモデルの違いを比べてみましょう

#### 生年月日と年齢を例にしてのドメインモデルの考え方

ドメインモデルでは 年齢が業務の関心ごとであれば年齢クラスを作ります。
年齢クラスは内部的に生年月日をインスタンス変数に保ちそのインスタンス変数を使って寝られを計算するロジックをメソッドとして持ちます 年齢を知りたいという関心事がありそれを計算するロジックの置き場所が必要だから年齢クラスを作るというアプローチです


#### 生年月日と年齢を例にしてのデータモデルの考え方

一方、データモデルでは年齢は記録すべきデータではありません。  
計算の結果です。
テーブルには計算のもとになる生年月日だけを記録します。   
手続き型の設計ではデータとロジックの整理を分けて考えるため年齢という関心ごとはデータモデルには登場しないのです

ドメインモデルは オブジェクト単位でプログラムを整理する技法です。  
業務データとその業務データを使う判断加工計算の業務ロジックをオブジェクトとして扱うたためのクラスが基本 プログラム単位です。  
それに対してデータモデルはどのようなデータを扱うことが業務上必要かを整理することとデータを使った判断かっこを計算のロジックを記述する機能の設計を別々に進めます。

手続き型アプローチの設計では全体を俯瞰し定義するところからスタートします段階的に分割しながらより詳細に定義を進めます その時に分割を同じレベルで進めることを重視します初期の段階では全ての粒度が大きく中間地点では全ての要素が10ぐらいの粒度になります最終的に全てをプログラミング可能な小さなりゅうとで定期しますいわゆるトップダウンのアプローチです分割のそれぞれの段階で全体的に網羅性や整合性を確認しながら詳細顔していくやり方です
オブジェクト指向のアプローチは部分に注目します9部品を作り始めそれを組み合わせながら段階的に全体竹を作っていきますボトムアップのアプローチです全体を定義し網羅性や整合性を重視しながら部分を設計していくとトップダウンのアプローチは データとロジックの一緒に考えるオブジェクト指向のやり方ではうまくいきません 全体を俯瞰し段階的に分解していくのはそれだけでも大変な作業ですその上データとロジックの詳細まで一緒に考えるのは負担が大きく非効率ですどうしても全体から部分を分割式たいのであればデータはデータ機能は昨日に分けて段階的に分割している方がまだやるやりやすいでしょう 一方部分から全体を組み立てよう区立ていくボトムアップのアプローチであればデータとロジックを一緒に考えることはむしろ自然ですある部分に注目してその部分に必要なデータとそのデータを使った判断学校設計のロジックだけを範囲を限定すればデータアートロジックを一緒に考えそのままクラスとして設定することは難しいことではありませんそうやって部分に注目してデータの業務ロジックを一体にして小さな部分ドメインオブジェクトを設定しながら全体を組み合わせていくのがオブジェクト指向のやり方です
