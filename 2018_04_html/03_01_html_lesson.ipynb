## CSSをHTMLに反映させるための指定方法について（classとid）

スタイルシートを記述する際、今回追記したh1のように「ある要素すべて（h1全体）」を一括して装飾するのではなく、「ある要素のうち一部だけ」を装飾したいときは class 属性と、id 属性を使用してCSSを反映します。  
※javascriptでも要素全体や一部を処理する場合はclassやidを多用します。

## idでCSSを適用してみる

それではidを利用してCSSを適用してみましょう。  
H2タグの文字を青色にして、画面の右へ表示させてみましょう。  
まずindex.htmlのH2タグにcssidという名前のidを設定します。  
文字の色を青色に、文字の表示位置を右端に設定をします。

## classでCSSを適用してみる
classは名前の前に.(ドット)をつけて記述をします。  
登録の際は文字の色を緑色で太字に設定をします。

例えば  
レース名[セントウルS]　のhrefの特徴には[race]と言うキーワード  
馬名[ファインドール]　のhrefの特徴には[horse]と言うキーワード  
騎手名[M.デム]　のhrefの特徴には[jocky]と言うキーワード  
調教師[高橋義忠]　のhrefの特徴には[trainer]と言うキーワード  
そのようなキーワードを元に持ってくる。

```python
from bs4 import BeautifulSoup 
import urllib
html = urllib.request.urlopen("http://race.netkeiba.com/?pid=special&id=0092")
soup = BeautifulSoup(html,"html5lib")

for a in soup.find_all('a'):
    p = a.get('href')
    if 'horse' in p:
        print(p)
    
```

下記のような馬の受賞歴をスクレーピングできたりします。  
処理が重たいので気をつけてください。  

```python
from bs4 import BeautifulSoup 
import urllib
import pandas as pd
html = urllib.request.urlopen("http://race.netkeiba.com/?pid=special&id=0092")
soup = BeautifulSoup(html,"html5lib")

# 某レースの出走馬のデータを取得
urls=[]
for a in soup.find_all('a'):
    p = a.get('href')
    if 'horse' in p:
        urls.append(p)
        
races = pd.DataFrame()

# 馬データからその戦績をまとめる
for url in urls:
    race = pd.read_html(url)
    wkrace = race[3]
    wkrace['uma'] = url[29:39]
    races =   pd.concat([races,wkrace])
races
```
