# World Wide Web(ワールドワイドウェブ)

インターネットは、少なくとも1960年代以降存在していた相互接続されたコンピュータネットワークのシステムです。  
しかし、インターネットの初期の使用は、情報を電子的に交換した少数の独立したネットワーク（大部分は政府、軍事、科学）に限られていました。   
1980年代には多くの研究機関や大学がインターネットを利用してデータを共有していました。  
ヨーロッパでは、スイスのジュネーブにあるCERN（欧州原子力研究機構）に最大のインターネットノードがあり、世界最大の粒子物理研究所を運営しています。  
これらの実験は膨大な量のデータを生成し、世界中の科学者と遠隔で共有する必要があります。

しかし今日と比較して、1980年代までの全体的なインターネット使用は非常に少なかった。ほとんどの人はそれにアクセスできなかったし、なぜそれが重要なのか理解していなかった。  
少数のインターネットノードがすべてのトラフィックに電力を供給し、それを使用するコンピュータは主に同じ小さなネットワーク内にありました。

これは、1989年にCERNの研究者、ティム・バーナーズ・リーがHTTPを発明し、現代のWWWで導かれたときに変更されました。彼の偉大な洞察は、既存のハイパーテキストシステムでは、コンピュータ画面に表示されるテキストが他のドキュメントへのリンク（ハイパーリンク）を含んでいて、インターネットに移動できるということでした。

彼の発明であるハイパーテキスト転送プロトコル（HTTP）は、インターネット上で文書を共有する最初の標準的な普遍的な方法でした。   
URL、リンク、イメージ、オーディオ、ビデオなどのリソースを持つ個別のドキュメントであるWebページという概念が導入されました。

今日、ほとんどの人が「インターネット」を考えるとき、彼らはワールドワイドウェブを考えています。これは現在、数十億人の人々とコンピュータがオンラインでコミュニケーションする主要な方法です。

# URL

URL（Uniform Resource Locator）は、インターネット上のリソースのアドレスです。たとえば、Googleのホームページはhttps:// www. google.comにあります。

Googleホームページにアクセスするには、完全なURLアドレスをWebブラウザに入力します。
あなたのブラウザはインターネットを介してリクエストを送信し、ブラウザにGoogleホームページを表示するために必要なデータを返すサーバーに、魔法のように接続します。

この要求と応答パターンは、すべてのWeb通信の基礎となります。  
クライアント（通常はウェブブラウザであるが、ネイティブアプリまたは実際にインターネットに接続されたデバイス）は情報を要求し、サーバはどこかで応答する。

ウェブ通信はHTTPを介して行われるので、これらはHTTPリクエストおよびHTTPレスポンスとして正式に知られている。

与えられたURL内には、いくつかの個別のコンポーネントもあります。  
たとえば、https:// www. google.comを再度ご検討ください。  
最初の部分httpsは使用されているスキームを参照します。 Webブラウザは、その場所のリソースにアクセスする方法を指示します。  
ウェブサイトの場合、これは通常httpまたはhttpsですが、ファイルの場合はftp、電子メールの場合はsmtpなどとすることもできます。  
次のセクション、www .google.comは、サイトのホスト名または実際の名前です。  
すべてのURLにスキームとホストが含まれています。

多くのWebページにはオプションのパスも含まれています。   
Pythonのホームページ（https:// www .python.org）で "About"ページのリンクをクリックすると、https:// www . python.org/ about/にリダイレクトされます。 / about / pieceはパスです。

要約すると、https:// python.org /about/のようなすべてのURLには、3つの潜在的な部分があります。

スキーム -  https

ホスト名 -  www. python.org

と（オプションの）パス -  / about /

インターネットプロトコルスイート

リソースの実際のURLを知ったら、クライアントとサーバーを接続して実際のWebページを読み込むために、他のテクノロジの集合全体が適切に（一緒に）動作している必要があります。  
これは広くインターネットprocotolスイートと呼ばれ、このトピックだけで書かれた全書籍があります。  
しかし、我々の目的のために、我々は幅広い基本に固執することができます。

ユーザーがhttps:// www. google.comをウェブブラウザに入力してヒットすると、いくつかのことが起こります。  
まず、ブラウザは膨大なインターネット上のどこかで、目的のサーバを見つける必要があります。  
ドメイン名サービス（DNS）を使用してドメイン名「google.com」をIPアドレスに変換します.IPアドレスは、インターネット上の接続されているすべてのデバイスを表す一連の数字です。  
ドメイン名は、 "2001：0db8：0000：0042：0000：8a2e：0370：7334"のようなIPアドレスよりも "google.com"のようなドメイン名を人間が覚えやすいため、使用されます。

ブラウザーに特定のドメインのIPアドレスが設定されたら、目的のサーバーとの一貫した接続を設定する方法が必要です。

これは、2つのアプリケーション間で信頼性の高い、順序付けされた、エラーチェックされたバイトの配信を提供するTCP（Transmission Control Protocol）を介して行われます。

2台のコンピュータ間でTCP接続を確立するには、クライアントとサーバーの間に3方向の「ハンドシェイク」が発生します。
 
1.クライアントは、接続を確立するように求めるSYNを送信します
2.サーバは、要求を確認し、接続パラメータを渡すSYN-ACKで応答します
3.クライアントは、接続を確認するACKをサーバーに返します

TCP接続が確立されると、2台のコンピュータはHTTP経由で通信を開始できます。

# HTTP

この章ではすでにHTTPについて多くのことを話しましたが、ここでは実際の内容とその動作について説明します。

HTTPは、既存のTCP接続を持つ2台のコンピュータ間の要求/応答プロトコルです。要求を行うコンピュータはクライアントと呼ばれ、コンピュータはサーバーと呼ばれます。  
通常、クライアントはウェブブラウザですが、iOSアプリまたは実際にインターネット接続されたデバイスでもかまいません。サーバーとは、インターネット上で動作するように最適化された任意のコンピュータの名前です。  
基本的なラップトップをサーバーに変えるために本当に必要なのは、特別なソフトウェアと永続的なインターネット接続だけです。

すべてのHTTPメッセージは、要求/ステータス行、ヘッダー、およびオプションの本文データで構成されています。たとえば、ブラウザがhttp：// www. google.comにあるGoogleホームページをリクエストするために送信するHTTPメッセージのサンプルを示します。

一番上の行はリクエスト行と呼ばれ、使用するHTTPメソッド（GET）、パス（/）、使用するHTTPのバージョン（HTTP / 1.1）を指定します。

その後の2つの行はHTTPヘッダーです。Hostはドメイン名で、Accept_Languageは使用する言語です（この場合はアメリカ英語です）。 利用可能なHTTPヘッダーが多数あります。

HTTPメッセージには、本体と呼ばれるオプションの第3セクションもあります。

ステータスコード

WebブラウザがURL上でHTTPリクエストを実行すると、実際に動作する保証はありません！ したがって、各HTTP応答に付随するかなり長いHTTPステータスコードのリストがあります。

以下のシステムに基づいて、ステータスコードの一般的なタイプを知ることができます。

- 2xx成功 - クライアントから要求されたアクションが受信され、理解され、受け入れられました

- 3xxリダイレクト - 要求されたURLが移動しました

- 4xxクライアントエラー  - エラーが発生しました。通常、クライアントからのURLリクエストが正しくありません。

- 5xxサーバーエラー  - サーバーが要求の解決に失敗しました

利用可能なステータスコードをすべて覚える必要はありません。  
練習では、200（OK）、201（作成）、301（常駐移動）、404（見つからない）、500（サーバーエラー）などの最も一般的なものに慣れるでしょう。

重要なことは、一般的に言えば、特定のHTTPリクエストには4つの潜在的な結果しかないということです。  
それは働いていました（2xx）、何とか（3xx）リダイレクトされた、クライアントがエラーを起こしました（4xx）エラー（5xx）。

これらのステータスコードは、すべてのHTTPメッセージの先頭にある要求/応答行に自動的に配置されます。

# Statelessness(無国籍)
最後の重要な点は、HTTPをステートレスなプロトコルにすることです。  
これは、各要求/応答のペアが前のものとは完全に独立していることを意味します。コンピュータサイエンスの状態として知られている過去のインタラクションの記憶された記憶はありません。

ステートレス性は、HTTPに多くの利点をもたらします。すべての電子通信システムは時間の経過とともに信号損失が発生するため、ステートレスプロトコルを使用していないと、1つの要求/応答サイクルが経過しなかった場合、事態が絶たれます。  
その結果、HTTPは非常に弾力のある分散プロトコルとして知られています。

しかし、欠点は、Webアプリケーションでは状態の管理が本当に重要であるということです。  
州は、あなたがログインしたこと、およびeコマースサイトがあなたのショッピングカートを管理する方法をウェブサイトがどのように記憶しているかを示します。現代のウェブサイトをどのように使用するかは基本的ですが、HTTP自体ではサポートされていません。

歴史的にはサーバー上では状態が維持されていましたが、React、Angular、Vueなどの最新のフロントエンドフレームワークでは、クライアント、Webブラウザにますます移動しました。ユーザー認証については、状態について詳しく説明しますが、HTTPはステートレスであることを覚えています。これは、2つのコンピュータ間で確実に情報を送信するのに非常に役立ちますが、個々の要求/応答のペアの外側にあるものはすべて覚えていません。

# REST

（Representational State Transfer）は、2000年にロイ・フィールディングが博士論文で提案したアーキテクチャです。  
これはWeb上にAPIを構築するアプローチです。  
これはHTTPプロトコルの上にあることを意味します。

APIを実際にRESTfulにするかどうかについては、全書籍が書かれています。  
しかし、ここでは3つの主な特徴があります。すべてのRESTful API：


- HTTPのようにステートレスです

- 一般的なHTTP動詞（GET、POST、PUT、DELETEなど）をサポートしています。
 
- JSON形式またはXML形式のデータを返します。

すべてのRESTful APIには、少なくとも次の3つの原則が必要です。この標準は、Web APIの設計と使用の両方に一貫性のある方法を提供するため、重要です。

# Conclusion(結論)

現代のワールドワイドウェブの基礎をなすテクノロジーはたくさんありますが、私たちは開発者としてすべてを最初から実装する必要はありません。 DjangoとDjango REST Frameworkの美しい組み合わせは、Web APIに関連する複雑さの大部分を適切に処理します。
